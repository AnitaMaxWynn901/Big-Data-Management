```python
from kafka import KafkaProducer
import json, time

producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda x: json.dumps(x).encode('utf-8')
)

events = [
    {"event": "Registration", "user": "User1", "time": "2025-04-10T10:00:00"},
    {"event": "Workshop", "user": "User2", "time": "2025-04-10T10:15:00"}
]

for event in events:
    producer.send('adt-events', value=event)
    print(f"Sent event: {event}")
    time.sleep(1)
```
